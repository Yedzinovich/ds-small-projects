### Part 3: Recommender Systems Bias
**Course:** DATA 612 – Recommender Systems<br>
**Student:** Inna Yedzinovich

Recommender systems definitely reinforce human bias. Based on what I was reading, there are several types of bias that show up in these systems, like popularity bias, exposure bias, position bias, and even demographic or algorithmic bias. Basically, the system keeps pushing certain items more than others, and users end up interacting with what is shown to them the most. This creates a kind of “bubble” where people mostly see similar content instead of a balanced view.

I’ve personally noticed this a lot, especially from my experience living in different countires. Recommendations in the US vs. Eastern Europe were very different — not just because of culture, but because systems push different products, movies, and lifestyles. Even now, I see how trends are shaped by recommendations, like being pushed toward “environmentally friendly” products or certain lifestyle brands just because that’s what’s popular or seen as “cool.” Over time, this can even change markets, like how thrift stores shifted from affordable places to more expensive “trendy” shops.

I also think recommender systems can definitely lead to unfair targeting and segmentation. People are often grouped based on behavior, demographics, or assumptions, and then shown different content, prices, or products. For example, things like higher-interest loans, the “pink tax,” or being shown more expensive products can all come from how users are categorized. I recently saw an example where different people using the same app (Instacart) were shown different prices for the same items at the same time. That’s a clear case of targeting and segmentation that can negatively affect fairness.

At the same time, I think these systems are not always bad. They can be helpful when used well — for example, recommending useful content, support communities, or movies that actually match someone’s interests. But the problem is that they are often optimized for engagement and sales rather than fairness. Personally, I notice this with ads — like after shopping during Black Friday, I keep getting targeted ads for similar items. Even if I like them, I don’t actually need them, but sometimes I still end up buying them.

Overall, I think recommender systems can both help and harm, but in practice they often reinforce bias because they rely on past behavior and push patterns that already exist. If not designed carefully, they can limit choices, influence decisions, and create unfair outcomes for different groups of users.
