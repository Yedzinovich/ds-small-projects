# Mitigating the Harm of Recommender Systems

Recommender systems today are mainly built around one goal — making money. In a capitalist system, platforms earn revenue from ads, which means they need users to spend as much time as possible watching content. To do that, they create highly personalized recommendations that keep people engaged, scrolling, watching, and interacting non-stop.

The problem is that this turns into overconsumption. People start watching too much content, especially short and fast-paced formats, and it affects how they think and behave. It becomes harder to focus, harder to enjoy simple things, and even finishing something like a full movie feels difficult. Many people develop bad habits — lack of sleep, irritation, and dependency on constant stimulation. Over time, this can affect attention, thinking, and overall well-being.

This is especially dangerous for younger users. Kids and teenagers are exposed to these systems early, and they can become “hooked” before they understand what is happening. Their behavior is shaped by constant recommendations, and small mistakes they make online can follow them for a long time. This can negatively impact their future, even though they are still learning and growing.

To reduce these harms, we need changes both in technology and in society. First, there should be stronger regulations to limit how addictive these systems can be, especially for younger users. Platforms should not design systems that are meant to keep people endlessly engaged. Features like autoplay and infinite scrolling should be limited. There should also be better protection for younger users, including restrictions on accounts and data collection.

Second, we as a society need to rethink how we treat online behavior. Young people should not be permanently punished for mistakes they made while growing up. And more broadly, we need to stop over-consuming content and return to more real-life interactions.

Recommender systems are powerful, but right now they are optimized for engagement, not well-being. If overused, they can affect the brain in ways similar to addictive behaviors, constantly pushing for more stimulation. To make things better, we need both better system design and a shift in how we use technology in everyday life.
