### Part 2: Recommendations at Scale with Spark
**Course:** DATA 612 – Recommender Systems<br>
**Student:** Inna Yedzinovich

In this talk, the most interesting part for me was how Spotify uses both mathematical techniques and real-world system design to build recommendations. In the first half, the focus was on the math, especially matrices and factorization using ALS. It’s been a while since I studied linear algebra, so it was interesting to see how these concepts are actually used in real applications. Once the idea behind ALS is understood, it becomes a powerful way to model user preferences. I also found the runtime comparisons interesting, since they showed how different implementations affect performance.

This also connects directly to what we did in Project 1. We started with a simple user-item matrix and used RMSE to evaluate predictions, which is basically the first step before scaling to more advanced systems like Spark. Seeing that connection helped me understand how these concepts grow into real-world systems.

In the second half, the focus shifted to data and system challenges. One thing that stood out to me was how slow Hadoop was compared to Spark, especially because ALS requires multiple iterations. Using Spark and in-memory computation significantly improves performance. It also highlighted how important system design is, not just the model itself.
