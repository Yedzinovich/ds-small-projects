### Part 1: Recommender System (Spotify)

Spotify is an application that recommends music to users. When a user creates an account, the system collects some basic information such as location and possibly demographic data. Based on this initial information, Spotify can already provide some early suggestions like playlists or popular songs.

As the user continues to use the app, Spotify tracks activity such as what songs are played the most, skipped, saved, added to playlists, or searched. Using this data, the system builds a profile of the user’s preferences and continuously improves recommendations. This approach appears to combine both content-based and collaborative filtering, since it uses both individual listening behavior and patterns from other users with similar tastes.

In my experience, Spotify provides very accurate recommendations for English music in the U.S. However, for international music, the recommendations sometimes feel more limited or less accurate.


### Part 2: Attacks on Recommender Systems

The article shows that people are often biased, and IMDb ratings do not always reflect the actual quality of a movie. Instead, ratings can be influenced by personal opinions about actors, cultural or political bias, or conflicts between groups. In this case, many users gave very low ratings to the movie before it was even released, which means the ratings were not based on real viewing experience.

A similar example can be seen on Amazon, where some sellers offer gift cards in exchange for 5-star reviews. I’ve also personally experienced something similar with dental offices asking for 5-star Google reviews in exchange for a Starbucks gift card, even when the service is overpriced or not very good. These examples show how easy it is to manipulate rating systems.

To prevent this kind of abuse, platforms should detect unusual rating patterns. For example, a large number of extreme ratings (only 1-star and 5-star) could be a red flag. Also, systems could require verification, such as proof that a user actually watched the movie (like a ticket purchase) before allowing them to leave a review. This would make ratings more reliable and reduce manipulation.
