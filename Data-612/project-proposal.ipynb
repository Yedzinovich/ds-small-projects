## Final Project Proposal: Large-Scale Video Game Recommendation System Using Apache Spark  

**Course:** DATA 612 – Recommender Systems<br>
**Student:** Inna Yedzinovich

## Introduction

For my final project, I plan to build a large-scale recommender system using Apache Spark and the Amazon Reviews 2023 Video Games dataset.

Unlike the datasets used in previous assignments, this dataset contains millions of user-item interactions, making it a good candidate for distributed processing. The goal is to generate personalized video game recommendations and explore how Spark can be used to handle larger recommendation problems efficiently.

## Dataset

The dataset comes from the Amazon Reviews 2023 collection and will use the **Video Games** category.

Dataset statistics:

- 2.8 million users
- 137 thousand items
- 4.6 million ratings

This exceeds the project requirement of working with a large dataset and provides enough user-item interactions for collaborative filtering and hybrid recommendation approaches.

## Proposed System

I plan to build a hybrid recommendation system consisting of:

1. **Collaborative Filtering**
   - Spark ALS (Alternating Least Squares)
   - Learns user preferences from rating behavior

2. **Content-Based Filtering**
   - Uses available product metadata and text information
   - Recommends similar games based on content features

3. **Hybrid Recommender**
   - Combines collaborative and content-based recommendations

The goal is to compare the different approaches and determine whether the hybrid model improves recommendation quality.

## Evaluation

I plan to evaluate the recommender system using a combination of metrics such as:

- RMSE
- Coverage
- Novelty
- Diversity

These metrics will help evaluate both recommendation accuracy and recommendation quality.

## Why Apache Spark?

This dataset contains millions of ratings and is significantly larger than the datasets used in earlier assignments. Apache Spark provides tools for distributed data processing and scalable machine learning, making it a suitable platform for building a recommender system at this scale.

I plan to use Databricks Community Edition for development and experimentation.

## Expected Outcome

The expected outcome is a scalable video game recommendation system capable of generating personalized recommendations for users. In addition to evaluating recommendation quality, I will compare the advantages and challenges of using Spark for building large-scale recommender systems.
